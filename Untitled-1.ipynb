**HYPOTHESIS TESTING**

NULL hypothesis: there is no difference in average fare between custoomers who use credit cards and customers who use cash.

Alternative hypotheses: there is a diffrence in average fare between customers who use credit cards and customers who use cash.

```python
sm.qqplot(df['fare_amount'], line='45')
plt.show()
```

```python
card_sample=df[df['payment_type'] == 'Credit Card']['fare_amount']
cash_sample=df[df['payment_type'] == 'Cash']['fare_amount']
```

```python
t_stat, p_value = stats.ttest_ind(card_sample, cash_sample, equal_var=False)
print(f"T-statistic: {t_stat}, P-value: {p_value}")
```

```python
df
```

```python
X = df[['duration']]  
y = df['fare_amount']     

model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

r2 = r2_score(y, y_pred)
mse = mean_squared_error(y, y_pred)

print("R-squared:", round(r2, 3))
print("Mean Squared Error:", round(mse, 2))
print("Regression Coefficient (slope):", round(model.coef_[0], 3))
print("Intercept:", round(model.intercept_, 3))

plt.figure(figsize=(10, 6))
sns.scatterplot(x='duration', y='fare_amount', data=df, alpha=0.3, label='Actual Data')
plt.plot(df['duration'], y_pred, color='red', label='Regression Line')
plt.title('Linear Regression: Trip Duration vs Fare Amount')
plt.xlabel('Trip Duration (minutes)')
plt.ylabel('Fare Amount')
plt.legend()
plt.grid(True)
plt.show()
```

We analyzed how trip duration affects fare using linear regression.
Results show that fare increases by approximately â‚¹0.69 per additional minute of the ride.
The model explains RÂ² = 0.76, meaning trip duration accounts for about 76% of the variation in fare amount.
This insight can help optimize pricing strategies, estimate trip costs more accurately, and detect irregular or unfair fare charges.
